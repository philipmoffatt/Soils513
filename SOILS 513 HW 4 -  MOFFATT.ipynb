## 1. (20 points) A solution contains 0.01 M CaCl2 at 20â€€C.
* (a) What is the osmotic pressure of this solution?
    - $\Pi$ = n$\gamma$C<sub>s</sub>RT
         
* (b) What is the relative humidity in the atmosphere above the solution and the water activity of this solution?
    - $\Pi$ = -($\rho$<sub>w</sub> R T ln(P/P<sub>o</sub>))/ M<sub>w</sub> 
    - a<sub>w</sub> = P/P<sub>o</sub> = exp ((-M<sub>w</sub> $\Pi$) / ($\rho$<sub>w</sub> R T))
    - RH = P/P<sub>o</sub> * 100

```python
import math
n = 3
gamma = 1 # assuming a dilute solution
C = 0.01 # mol/L
C = C*998 # M/m^3 assuming density of water is 998 kg/m^3
T = 293 # Kelvin
R = 8.3144 # J/mol K
potO = - n*C*R*T
print('Osmotic Potential =',potO/100,'hPa')
```

```python
Mw = 0.018 #kg/mol
aw = math.exp((Mw*potO)/(998*R*T))
print('water activity is',aw)
print('RH =',aw*100,'%')
```

## 2. (20 points) A solution contains 0.01 M NaBr and 0.03 M CaCl2 at 20â€€C.
- (a) What is the osmotic pressure of this solution?
    - $\Pi$<sub>NaBr</sub> = nCRT
    - $\Pi$<sub>tot</sub> = $\Pi$<sub>NaBr</sub> + $\Pi$<sub>CaCl2</sub>
- (b) What is the relative humidity in the atmosphere above the solution and the water activity of this solution?

```python
n = 2
gamma = 1 # assuming a dilute solution
C = 0.01 # mol/L
C = C*998 # M/m^3 assuming density of water is 998 kg/m^3
T = 293 # Kelvin
R = 8.3144 # J/mol K
potNaBr = - n*C*R*T
print('NaBr Osmotic Potential =',potNaBr,'Pa')
n = 3
gamma = 1 # assuming a dilute solution
C = 0.03 # mol/L
C = C*998 # M/m^3 assuming density of water is 998 kg/m^3
T = 293 # Kelvin
R = 8.3144 # J/mol K
potCaCl2 = - n*C*R*T
potTot = potCaCl2 + potNaBr
print('CaCl2 Osmotic Potential =',potCaCl2,'Pa')
print('Total Osmotic Potential = ',(potTot)/100, 'hPa')
```

```python
aw = math.exp((Mw*potTot)/(998*R*T))
print('water activity is',aw)
print('RH = ',aw*100,'%')
```

## 4. (20 points) Consider a capillary tube inserted in a reservoir of water.
(a) What is the relative humidity at equilibrium at 300 K above the meniscus in a capillary tube of
radius r = 0.1 â€€micrometers if the contact angle is â€€ = 0?

(b) If the contact angle changes, does the relative humidity remain the same or does it change? If it
changes, would the relative humidity increase or decrease?

    - The RH does change in relation to the contact angle. It appears to increase, indicating more water present in the gas phase above the meniscus; see calculations for a=45 degrees below.

```python
T = 300 # kelvin
r = 0.0000001 # m
a = 0 # degrees
rhoW = 996 # kg/m^3
Mw = 0.018 # kg/mol
R = 8.3144 # J/mol K
g = 9.81 # m/s^2
sigma = 0.072 # N or J/m^2
h = 2*sigma*math.cos(math.radians(a))/(rhoW*g*r)
print('capillary rise (a=0)=',h,'m')
RH = 100*math.exp((-h*g*rhoW*Mw)/(rhoW*R*T))
print('Relative Humidity (a=0)= ',RH, '%')

a = 45 # degrees, this is a change to the shape of the meniscus
h = 2*sigma*math.cos(math.radians(a))/(rhoW*g*r)
print('capillary rise (a=45)=',h,'m')
RH = 100*math.exp((-h*g*rhoW*Mw)/(rhoW*R*T))
print('Relative Humidity (a=45)= ',RH, '%')
```
